`JOIN` - Оператор для соединения таблиц

СИНТАКСИС

```sql
SELECT
        столбец1,
	столбец2,
	...
FROM
	таблица1
JOIN таблица2 ON условие
JOIN таблица3 ON условие
```
Порядок присоединения таблиц в **данном случае не важен** — результат будет одинаковым.

```sql
SELECT 
    long_name,
    home_team_goals,
    away_team_goals
FROM
    sql.teams
JOIN sql.matches on home_team_api_id = api_id
```

```sql
SELECT
        столбец1,
	столбец2,
	...
FROM
	таблица1
JOIN таблица2 ON условие
JOIN таблица3 ON условие
```
